# 2. Data processing

```python
X_drop=X.drop(['Over18','StandardHours'],1)
```

## 2.2 Deal with the NA data

```python
print('There are %d rows which have NA data' %len(X_drop.loc[X_drop.isnull().any(axis=1)]))
```

# 3. Model selection

## 3.0 Load data 

```python
X_d_o_n=pd.read_csv('data_after_procession.csv')
y=pd.read_csv('class_after_procession.csv')['Attrition']
```

## 3.1 Deal with the imbalance data

```python
#show the class imbalance
y.hist()
plt.title('Number of different class')
plt.show()

#deal  with the imbalance with resampling method
from sklearn.utils import resample
print ('Old data, Num of class0 & class 1 are %.0f ,%.0f' % (sum(y==0),sum(y==1)))
X_upsampled,y_upsampled=resample(X_d_o_n[y==1],y[y==1],replace=True,n_samples=sum(y==0),random_state=123)

X_bal=np.vstack((X_d_o_n[y==0],X_upsampled))
y_bal=np.hstack((y[y==0],y_upsampled))

print ('New data, Num of class0 & class 1 are %.0f ,%.0f' % (sum(y_bal==0),sum(y_bal==1)))
```

## 3.2 Data split
need 0.01s

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X_bal,y_bal,test_size=0.3,random_state=1,stratify=y_bal)

#change the datatype of X
X_train=X_train.astype(np.float)
X_test=X_test.astype(np.float)
print('Labels counts in y:', np.bincount(y_bal))
print('Labels counts in y_train:', np.bincount(y_train))
print('Labels counts in y_test:', np.bincount(y_test))

```

## 3.3 Standardize the features 
need 0.01 s

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(X_train)
X_train_std=sc.transform(X_train)
X_test_std=sc.transform(X_test)
```

## 3.4 Do the PCA  
need 1 s

```python
#define the function  which caculates the components num we need to meet the target variance contribution
def get_target_n(X_train_std,X_test_std,taget_contribution=0.95):
    from sklearn.decomposition import PCA
    pca = PCA()

    X_train_pca = pca.fit_transform(X_train_std)
    X_test_pca = pca.transform(X_test_std)
    total_contribution=pca.explained_variance_ratio_
    target_contribution=np.cumsum(total_contribution)
    n_need=len(target_contribution[target_contribution<=taget_contribution])
    print('To get target variance contribution %.0f %%, we need %.0f components.'%(taget_contribution*100,n_need))
    print('')
    return n_need
```

```python
from sklearn.decomposition import PCA
n_need=get_target_n(X_train_std,X_test_std,taget_contribution=0.9)
pca = PCA(n_components=n_need)
X_train_pca = pca.fit_transform(X_train_std)
X_test_pca = pca.transform(X_test_std)

num= len(pca.explained_variance_ratio_)+1
plt.bar(range(1,num), pca.explained_variance_ratio_, alpha=0.5, align='center')
plt.step(range(1, num), np.cumsum(pca.explained_variance_ratio_), where='mid')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.title('Cumulative variance contribution')
plt.show()
```

## 3.6 Draw the ROC curve
need 10 s

```python
t0=tt()


from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
#define the best 4 classifier
lr = LogisticRegression(C=100,random_state=1)
lr.fit(X_train_pca, y_train)

svm = SVC(kernel='linear', C=1, random_state=1,probability=True)
svm.fit(X_train_pca, y_train)

knn = KNeighborsClassifier(n_neighbors=1, p=2,metric='minkowski')
knn.fit(X_train_pca, y_train)

tree = DecisionTreeClassifier(criterion='gini', max_depth=4, random_state=1)
tree.fit(X_train_pca, y_train)

all_clf=[lr,svm,knn,tree]
clf_labels=['LogisticRegression','SVM','K_means','Decision tree']


#Draw the ROC curve of 4 classfier
from sklearn.metrics import roc_curve
from sklearn.metrics import auc

colors = ['black', 'orange', 'blue', 'green']
linestyles = [':', '--', '-.', '-']
for clf, label, clr, ls \
        in zip(all_clf,
               clf_labels, colors, linestyles):

    # assuming the label of the positive class is 1
    y_pred = clf.fit(X_train_pca,
                     y_train).predict_proba(X_test_pca)[:, 1]
    fpr, tpr, thresholds = roc_curve(y_true=y_test,
                                     y_score=y_pred)
    roc_auc = auc(x=fpr, y=tpr)
    plt.plot(fpr, tpr,
             color=clr,
             linestyle=ls,
             label='%s (auc = %0.2f)' % (label, roc_auc))

    
plt.legend(loc='lower right')
plt.plot([0, 1], [0, 1],
         linestyle='--',
         color='gray',
         linewidth=2)
plt.xlim([-0.1, 1.1])
plt.ylim([-0.1, 1.1])
plt.grid(alpha=0.5)
plt.title('The ROC curve of 4 classifier')
plt.xlabel('False positive rate (FPR)')
plt.ylabel('True positive rate (TPR)')



#plt.savefig('images/07_04', dpi=300)
plt.show()

t1=tt()
print('#-------------------------------------------------------------------#')
print (' The time cost of this step is %.02f s'%(t1-t0))
print('#-------------------------------------------------------------------#')
```

# 4. Improvement

## 4.1 Show the disadvantage of PCA

### 4.1.1 Define plot function

```python
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt


def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):

    # setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], 
                    y=X[y == cl, 1],
                    alpha=0.8, 
                    c=colors[idx],
                    marker=markers[idx], 
                    label=cl, 
                    edgecolor='black')

    # highlight test samples
    if test_idx:
        # plot all samples
        X_test, y_test = X[test_idx, :], y[test_idx]

        plt.scatter(X_test[:, 0],
                    X_test[:, 1],
                    c='',
                    edgecolor='black',
                    alpha=1.0,
                    linewidth=1,
                    marker='o',
                    s=100, 
                    label='test set')
```

### 4.1.2 Take Logistic Regression as an example

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(C=100, random_state=1,penalty='l2')
lr.fit(X_train_pca[:,:2], y_train)

X_combined_std = np.vstack((X_test_pca[:,:2],X_train_pca[:,:2]))
y_combined = np.hstack((y_test,y_train))

plot_decision_regions(X=X_combined_std, y=y_combined,
                      classifier=lr, test_idx=range(len(y_test)))

plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend(loc='upper left')
plt.title('Classification under PC1,PC2- logistic regression')
plt.tight_layout()
#plt.savefig('images/03_01.png', dpi=300)
plt.show()

print('The principles can not seperate the data linearly.')
```

## 4.2 Do the Kernel PCA to seperate the data in high dimention

## 4.4 Test model stability by k-fold
need 110s

```python
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import KernelPCA
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

print('K-fold-test')
# test Logistic regression
print('Begin to test Logistic regression...')
t0=tt()
pipe_lr= make_pipeline(StandardScaler(),KernelPCA(kernel='rbf',gamma=15,n_components=400),LogisticRegression(C=100, random_state=1))
scores=cross_val_score(estimator=pipe_lr,X=X_train,y=y_train,cv=10,n_jobs=-1)
t1=tt()
print('The accuracy of Logistic regression is %.2f+/-%.2f   , the time cost is %.02f s'%(np.mean(scores),np.std(scores),t1-t0))

# test k_means
print('Begin to test k_means...')
t0=tt()
pipe_knn= make_pipeline(StandardScaler(),KernelPCA(kernel='rbf',gamma=15,n_components=400),KNeighborsClassifier(n_neighbors=1,p=2, metric='minkowski'))  
scores=cross_val_score(estimator=pipe_knn,X=X_train,y=y_train,cv=10,n_jobs=-1)
t1=tt()
print('The accuracy of K_mean is %.2f+/-%.2f   , the time cost is %.02f s'%(np.mean(scores),np.std(scores),t1-t0))

# test tree
print('Begin to test tree...')
t0=tt()
pipe_tree= make_pipeline(StandardScaler(),KernelPCA(kernel='rbf',gamma=15,n_components=400),DecisionTreeClassifier(criterion='gini', max_depth=4,random_state=1))  
scores=cross_val_score(estimator=pipe_tree,X=X_train,y=y_train,cv=10,n_jobs=-1)
t1=tt()
print('The accuracy of Tree is %.2f+/-%.2f   , the time cost is %.02f s'%(np.mean(scores),np.std(scores),t1-t0))

# test SVM
print('Begin to test SVM...')
t0=tt()
pipe_svm= make_pipeline(StandardScaler(),KernelPCA(kernel='rbf',gamma=15,n_components=400),SVC(kernel='linear', C=100, random_state=1))  
scores=cross_val_score(estimator=pipe_svm,X=X_train,y=y_train,cv=10,n_jobs=-1)
t1=tt()
print('The accuracy of SVM is %.2f+/-%.2f   , the time cost is %.02f s'%(np.mean(scores),np.std(scores),t1-t0))
```

## 4.5 Draw the ROC curve
need 10 s

```python
t0=tt()

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

#select the suitable number of conponents after kpca.
feature_need=400

#define the 4 classifier
lr = LogisticRegression(C=100,random_state=1)
lr.fit(X_train_kpca[:,0:feature_need], y_train)

svm = SVC(kernel='linear', C=100, random_state=1,probability=True)
svm.fit(X_train_kpca[:,0:feature_need], y_train)

knn = KNeighborsClassifier(n_neighbors=1, p=2,metric='minkowski')
knn.fit(X_train_kpca[:,0:feature_need], y_train)

tree = DecisionTreeClassifier(criterion='gini', max_depth=4, random_state=1)
tree.fit(X_train_kpca[:,0:feature_need], y_train)

all_clf=[lr,svm,knn,tree]
clf_labels=['LogisticRegression','SVM','K_means','Decision tree']


#draw the ROC curve
from sklearn.metrics import roc_curve
from sklearn.metrics import auc

colors = ['black', 'orange', 'blue', 'green']
linestyles = [':', '--', '-.', '-']
for clf, label, clr, ls \
        in zip(all_clf,
               clf_labels, colors, linestyles):

    # assuming the label of the positive class is 1
    y_pred = clf.fit(X_train_kpca[:,0:feature_need],
                     y_train).predict_proba(X_test_kpca[:,:feature_need])[:, 1]
    fpr, tpr, thresholds = roc_curve(y_true=y_test,
                                     y_score=y_pred)
    roc_auc = auc(x=fpr, y=tpr)
    plt.plot(fpr, tpr,
             color=clr,
             linestyle=ls,
             label='%s (auc = %0.2f)' % (label, roc_auc))

plt.legend(loc='lower right')
plt.plot([0, 1], [0, 1],
         linestyle='--',
         color='gray',
         linewidth=2)

plt.xlim([-0.1, 1.1])
plt.ylim([-0.1, 1.1])
plt.grid(alpha=0.5)
plt.title('The ROC curve of 4 classifier after kpca')
plt.xlabel('False positive rate (FPR)')
plt.ylabel('True positive rate (TPR)')

#plt.savefig('images/07_04', dpi=300)
plt.show()



t1=tt()
print('#-------------------------------------------------------------------#')
print (' The time cost of this step is %.02f s'%(t1-t0))
print('#-------------------------------------------------------------------#')
```
