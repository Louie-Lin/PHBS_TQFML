```python
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt


def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):

    # setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], 
                    y=X[y == cl, 1],
                    alpha=0.8, 
                    c=colors[idx],
                    marker=markers[idx], 
                    label=cl, 
                    edgecolor='black')

    # highlight test samples
    if test_idx:
        # plot all samples
        X_test, y_test = X[test_idx, :], y[test_idx]

        plt.scatter(X_test[:, 0],
                    X_test[:, 1],
                    c='',
                    edgecolor='black',
                    alpha=1.0,
                    linewidth=1,
                    marker='o',
                    s=100, 
                    label='test set')
```

<font face="微软雅黑" size=4> First, we change the month to number from 1 to 12  <br />
 Second, we pick iteratively two features of data <br />
&emsp; &emsp; 1. Let y be the target vector, X be the dataof this two features<br />
&emsp; &emsp; 2. Take one feature of X and judge whether this feature is nominal categorical feature. <br />
&emsp; &emsp; 3. If yes, judge wheather this feature only have 'yes'or 'no'. If yes, turn 'yes' to 1, 'no' to 0; <br />
&emsp; &emsp; &emsp;Otherwise, turn this feature into dummies variables directly. <br />
 &emsp; &emsp; 4. Split X,Y into training and test set and fit model with training data ;<br /> 
 &emsp; &emsp; 5. Get prediction by applying the model to the test set, and calculate the accuracy ; <br />
 &emsp; &emsp; 6. Compare the accuracy with the best accuracy which get before, if better, update the best accuracy  ; <br />
 Finally, we print out the best 2 features and the best accuracy  ; <br />
 If the best 2 features do not turn into dummies features, plot the classification of test set.  ; <br /> </font>

```python
def find_best_feature_ppn(data):
    month_list=['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
    month_mapping=dict(zip(month_list,range(1,13)))
    data.loc[:,'month']=data.loc[:,'month'].map(month_mapping)
    basic_set=data.columns[:-1]
    best_feature=[]
    best_feature_with_dummies=[]
    best_accuracy=0
    x_train_best=0
    x_test_best=0
    y_train_best=0
    y_test_best=0
    
    for i in range(len(basic_set)-1):
        for j in range(i+1,len(basic_set)):
            feature_tem=basic_set[[i,j]]
            yes_no_mapping={'yes':1,'no':0}
            y=data.y
            y=y.map(yes_no_mapping).values
            
            X=data.loc[:,feature_tem]
            indicator_str=[]
            feature=[]
            feature_with_dummies=[]
            for name in feature_tem:
                if isinstance(X.loc[:,name].values[0], str):
                    feature_list=list(set(X.loc[:,name].values))
                    if len(feature_list)==2:
                        X.loc[:,name]=X.loc[:,name].map(yes_no_mapping)
            X=pd.get_dummies(X[X.columns])
            feature_with_dummies_tem=X.columns
            X=X.values.astype(float)
            ##split
            X_train, X_test, y_train, y_test = train_test_split(
                X, y, test_size=0.3, random_state=1, stratify=y)
            ##std
            sc = StandardScaler()
            sc.fit(X_train)
            X_train_std = sc.transform(X_train)
            X_test_std = sc.transform(X_test)

                    
            from sklearn.linear_model import Perceptron
            ppn = Perceptron(n_iter=40, eta0=0.1, random_state=1)
            ppn.fit(X_train_std, y_train)
            y_pred = ppn.predict(X_test_std)
            from sklearn.metrics import accuracy_score
            new_accuracy=accuracy_score(y_test, y_pred)
            if new_accuracy>best_accuracy:
                best_accuracy=new_accuracy
                best_feature=feature_tem
                best_feature_with_dummies=feature_with_dummies_tem
                x_train_best=X_train_std
                x_test_best=X_test_std
                y_test_best=y_test
                y_train_best=y_train
    print('Best Accuracy: %.2f' % best_accuracy)
    print('Best_feature:',best_feature)
    print('Best_feature_with_dummies:',best_feature_with_dummies)
    print('Num Best_feature_with_dummies:',len(best_feature_with_dummies))
    if len(best_feature_with_dummies)==2:
        ppn.fit(x_train_best,y_train_best)

        plot_decision_regions(X=x_test_best, 
                          y=y_test_best,
                          classifier=ppn)            
        plt.xlabel(best_feature[0])
        plt.ylabel(best_feature[1])
        plt.legend(loc='upper left')
        plt.tight_layout()
        #plt.savefig('images/03_06.png', dpi=300)
        plt.show()     
    #return([x_train_best,x_test_best,y_train_best,y_test_best,best_feature,best_feature_with_dummies])   
    else:
        print('Too much feature, cannot plot!')
           

                
            
            
```

```python
find_best_feature_ppn(data.copy())
```

```python
class LogisticRegressionGD(object):
    """Logistic Regression Classifier using gradient descent.

    Parameters
    ------------
    eta : float
      Learning rate (between 0.0 and 1.0)
    n_iter : int
      Passes over the training dataset.
    random_state : int
      Random number generator seed for random weight
      initialization.


    Attributes
    -----------
    w_ : 1d-array
      Weights after fitting.
    cost_ : list
      Sum-of-squares cost function value in each epoch.

    """
    def __init__(self, eta=0.05, n_iter=100, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state

    def fit(self, X, y):
        """ Fit training data.

        Parameters
        ----------
        X : {array-like}, shape = [n_samples, n_features]
          Training vectors, where n_samples is the number of samples and
          n_features is the number of features.
        y : array-like, shape = [n_samples]
          Target values.

        Returns
        -------
        self : object

        """
        rgen = np.random.RandomState(self.random_state)
        self.w_ = rgen.normal(loc=0.0, scale=0.01, size=1 + X.shape[1])
        self.cost_ = []

        for i in range(self.n_iter):
            net_input = self.net_input(X)
            output = self.activation(net_input)
            errors = (y - output)
            self.w_[1:] += self.eta * X.T.dot(errors)
            self.w_[0] += self.eta * errors.sum()
            
            # note that we compute the logistic `cost` now
            # instead of the sum of squared errors cost
            cost = -y.dot(np.log(output)) - ((1 - y).dot(np.log(1 - output)))
            self.cost_.append(cost)
        return self
    
    def net_input(self, X):
        """Calculate net input"""
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def activation(self, z):
        """Compute logistic sigmoid activation"""
        return 1. / (1. + np.exp(-np.clip(z, -250, 250)))

    def predict(self, X):
        """Return class label after unit step"""
        return np.where(self.net_input(X) >= 0.0, 1, 0)
        # equivalent to:
        # return np.where(self.activation(self.net_input(X)) >= 0.5, 1, 0)
```

```python
def find_best_feature_logit(data):
    month_list=['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
    month_mapping=dict(zip(month_list,range(1,13)))
    data.loc[:,'month']=data.loc[:,'month'].map(month_mapping)
    basic_set=data.columns[:-1]
    best_feature=[]
    best_feature_with_dummies=[]
    best_accuracy=0
    x_train_best=0
    x_test_best=0
    y_train_best=0
    y_test_best=0
    
    for i in range(len(basic_set)-1):
        for j in range(i+1,len(basic_set)):
            feature_tem=basic_set[[i,j]]
            yes_no_mapping={'yes':1,'no':0}
            y=data.y
            y=y.map(yes_no_mapping).values
            
            X=data.loc[:,feature_tem]
            indicator_str=[]
            feature=[]
            feature_with_dummies=[]
            for name in feature_tem:
                if isinstance(X.loc[:,name].values[0], str):
                    feature_list=list(set(X.loc[:,name].values))
                    if len(feature_list)==2:
                        X.loc[:,name]=X.loc[:,name].map(yes_no_mapping)
            X=pd.get_dummies(X[X.columns])
            feature_with_dummies_tem=X.columns
            X=X.values.astype(float)
            ##split
            X_train, X_test, y_train, y_test = train_test_split(
                X, y, test_size=0.3, random_state=1, stratify=y)
            ##std
            sc = StandardScaler()
            sc.fit(X_train)
            X_train_std = sc.transform(X_train)
            X_test_std = sc.transform(X_test)

                    
            lrgd = LogisticRegressionGD(eta=0.05, n_iter=1000, random_state=1)
            lrgd.fit(X_train_std, y_train)
            y_pred = lrgd.predict(X_test_std)
            from sklearn.metrics import accuracy_score
            new_accuracy=accuracy_score(y_test, y_pred)
            if new_accuracy>best_accuracy:
                best_accuracy=new_accuracy
                best_feature=feature_tem
                best_feature_with_dummies=feature_with_dummies_tem
                x_train_best=X_train_std
                x_test_best=X_test_std
                y_test_best=y_test
                y_train_best=y_train
    print('Best Accuracy: %.2f' % best_accuracy)
    print('Best_feature:',best_feature)
    print('Best_feature_with_dummies:',best_feature_with_dummies)
    print('Num Best_feature_with_dummies:',len(best_feature_with_dummies))
    if len(best_feature_with_dummies)==2:
        lrgd.fit(x_train_best,y_train_best)
        
        plot_decision_regions(X=x_test_best, 
                          y=y_test_best,
                          classifier=lrgd)            
        plt.xlabel(best_feature[0])
        plt.ylabel(best_feature[1])
        plt.legend(loc='upper left')
        plt.tight_layout()
        #plt.savefig('images/03_06.png', dpi=300)
        plt.show()     
    #return([x_train_best,x_test_best,y_train_best,y_test_best,best_feature,best_feature_with_dummies])   
    else:
        print('Too much feature, cannot plot!')
           

                
            
            
```
